**PROPERTIES OF NDARRAY**

```python
import numpy as np
import sys

def arrayProp():
  l1=[10,20,30,40,50]
  print(f"{l1} ,type: {type(l1)}")
  print(f"size of one element in list: {sys.getsizeof(l1[0])}")
  print(f"Length of l1:{len(l1)}")
  print(f"size of l1 :{sys.getsizeof(l1)*len(l1)}")

  print("-*-*"*20)

  a1=np.array([10,20,30,40,50])
  print(f"{a1} ,type: {type(a1)}")
  #how much memory is allocated to each and every element in ndarray
  print(f"size of one item: {a1.itemsize} bytes")

  #length of a1
  print(f"length of a1: {a1.size} ")

  #size of a1
  print(f"size of a1: {a1.size*a1.itemsize}")

  print("-*-*"*20)
  a2=np.array([10,20,30,40,50],dtype=np.int8)
  print(f"{a2} ,type: {type(a2)}")
  print(f"size of one item: {a2.itemsize} bytes")
  print(f"length of a2: {a2.size} ")
  print(f"size of a2: {a2.size*a2.itemsize}")

  print("-*-*"*20)
  a3=np.array([10,20,30,40,50],dtype=np.int16)
  print(f"{a3} ,type: {type(a3)}")
  print(f"size of one item: {a3.itemsize} bytes")
  print(f"length of a3: {a3.size} ")
  print(f"size of a3: {a3.size*a3.itemsize}")

arrayProp()
```

**TO SHOW NUMPY ARRAY TAKES LESSER TIME TO PROCESS DATA ⬇️**

```python
import time
count=1000
def func1():
  l1=list(range(count))
  l2=list(range(count))
  l3=[]
  t1=time.time()
  for index in range(count):
    l3.append(l1[index]+l2[index])
  t2=time.time()
  print(f"Time Taken by list to process: {t2-t1}  ms")

func1()
```

**FILTERING**

*   Combination of indexing and broadcasting




```python
def fun4():
  a1=np.array([10000,15000,120000,85000,150000,41000,90000])
  print(f"a1={a1}")

  #print(f"Emp having salary greater than 100000 : {a1 > 100000}") HERE we were getting only True False Values to get the actual values we use filtering

  #Filtering
  print(f"Emp having salary greater than 100000 :{a1[a1>100000]}")
  print("-*-"*20)
  #Ex:How many people are eligible for voting
  a2=np.array([39,19,1,4,9,56,45,60,65])
  print(f"People in family which are Eligible for Voting: {a2[a2>=18]}")
  print(f"People in family which are not Eligible for Voting: {a2[a2<18]}")
  print("-*-"*20)
  #can be done using filter
  a3=list(filter(lambda x:x>=18,a2))
  print(f"People in family which are Eligible for Voting: {a3}")
  print("-*-"*20)
fun4()
```

#iloc

```python
def function9():
  cars=[
      {"model":"i10","price":7.5,"company":"Hyundai"},
      {"model":"i20","price":10,"company":"Hyundai"},
      {"model":"Nano","price":2.5,"company":"tata"}
  ]
  df=pd.DataFrame(cars,columns=["model","price","company"])
  print(df)

  print("-*-"*20)
  #we use the concept of iloc to get the value at i'th position
  #print(df[0]) it will give keyerror no such a key

  #we will use iloc
  print(df.iloc[0,0])

  #row=0 column=0
  print(df.iloc[0,0])
  print("-*-"*20)

  #row=0 column=2
  print(df.iloc[0,2])
  print("-*-"*20)

  #row=0 column=[0,1]  #it will give 0th row and 0,1 columns
  print(df.iloc[0,0:2])
  print("-*-"*20)

  #row=[0:3] ,column=1  #it will give 0,1,2th row and 1 columns
  print(df.iloc[0:4,1])
  #print(df.iloc[0:,1])
  #print(df.iloc[:,1])
  print("-*-"*20)

  #to get all the rows and all the columns
  print(df.iloc[:,:])

function9()
```

```python
def function9():
  cars=[
      {"model":"i10","price":7.5,"company":"Hyundai"},
      {"model":"i20","price":10,"company":"Hyundai"},
      {"model":"Nano","price":2.5,"company":"tata"}
  ]
  df=pd.DataFrame(cars,columns=["model","price","company"])
  print(df)

  print("-*-"*20)
  #filtering
  #all the affordable cars
  print(df[df.price<10])

  #now we want to get all the rows and only model column and which are affordable
  #we can combine this two things
  #its like: select * from cars where price<10
  print(df.iloc[:,1][df.price<10])
function9()
```
