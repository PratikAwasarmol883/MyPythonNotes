AN EXCEPTION IS A SCENARION BECAUSE OF WHICH APPLICATION MAY CRASH.IT IS DYNAMIC OR RUNTIME .

```python
def divide(p1,p2):
  try:
    divi= p1/p2
    print(divi)
  except:     #generic except block
    print("An Exception is Raised.")
divide(10,0)
```

```python
class InvalidAgeError(Exception):
  pass
class InvalidSalaryError(Exception):
  pass

def person_info():
  name=input("Enter Name: ")
  age=int(input("Enter Age: "))
  salary=float(input("Enter Salary: "))
  if age>0:
    print(f"Age :{age}")
  else:
    raise InvalidAgeError()
  if salary>0:
    raise InvalidSalaryError()

try:
  person_info()
except InvalidAgeError as e:
  print(f"Invalid Age Exception: {e}")
except InvalidSalaryError as e:
  print(f"Invalid salary Exception: {e}")
```

**GENERIC EXCEPTION HAS TO BE ATLAST OF EXCEPTION**

```python
def divide(p1,p2):
  try:
    divi= p1/p2
  except ZeroDivisionError:     #Specific except block
    print("An Exception is Raised: Second Parammeter is Zero'")
  else:
    print(divi)

#divide(10,0)
divide(10,20)
```

#FINALLY BLOCK WILL EXECUTE WHEATHER EXCEPTION IS RAISED OR NOT.IT WILL ALWAYS EXCUTE.

```python
def fun():
  try:
    file=open("f","r")
  except Exception as e:
    print(e)
  finally:
    print("File Closed")

fun()
```
