#USING PRIVATE MEMBERS ⬇️

**Private members wont be able to access through reference of an object.**

```python
class Car:
  def __init__(self,model,company,price):
    #Private Members
    #name mangling=compiler changes private members name
    self.__model=model
    self.__company=company
    self.__price=price
  def print_info(self):
    print(f"Model:{self.__model}")
    print(f"Company:{self.__company}")
    print(f"Price in Lakhs:{self.__price}")
cars_1=Car("Supra","Toyoto",55)
cars_1.print_info()
print(cars_1.__dict__)
cars_1.price="Invalid Value"  #value will not be changed
cars_1.print_info()

cars_1._Car__price=10   #here we are able to tamper private member but it is not recommended
cars_1.print_info()
```

**ALTHOUGH WE ARE USING PRIVATE MEMBERS WE SHOULD HAVE FLEXIBILITY TO CHANGE THEM IN FUTURE FOR THAT WE WILL USE METHODS.**

```python
class Car:
  #initializer
  def __init__(self,model,company,price):
    print("Inside Initializer")
    #Private Members
    #name mangling=compiler changes private members name
    self.__model=model
    self.__company=company
    self.__price=price

  #de-initializer
  def __del__(self):
    print("Inside De-Initializer")

  #setter or Mutator
  def setprice(self,price):
    if price>0:
      self.__price=price
    else:
      print("Invalid Value")

  #setter or Mutator
  def setmodel(self,model):
    self.__model=model

  #setter or Mutator
  def setcompany(self,company):
    self.__company=company

  #getter or inspector
  def getcompany(self):
    return self.__company

  #getter or inspector
  def getprice(self):
    return self.__price

  #getter or inspector           using we are able to access single values like price,model,company
  def getmodel(self):
    return self.__model

  #facilitator
  def print_info(self):
    print(f"Model:{self.__model}")
    print(f"Company:{self.__company}")
    print(f"Price in Lakhs:{self.__price}")

cars_1=Car("Supra","Toyoto",55)
cars_1.print_info()
print("-*-"*20)
print(cars_1.__dict__)
print("-*-"*20)
#Now to change the value of Private member Price we will use
cars_1.setprice(85)
cars_1.print_info()

cars_1.getmodel()

#delete the memory
del cars_1
```

Single Inheritance Example

```python
class Person:
  def __init__(self,name):
    #Protected (using single underscore)
    #Accessible inside the same class , derived class
    self._name=name
  def print_info(self):
    print(f"name :{self.name}")

class Player(Person):
  def __init__(self,name,team):
    Person.__init__(self,name)
    self.team=team
  def print_player_info(self):
    print(f"team :{self.team}")
    print(f"Player :{self._name}")
```

Multi-level Inheritance

```python
class Person:
  def __init__(self,name):
    self._name=name    #protected
class Employee(Person):
  def __init__(self,name,empid):
    Person.__init__(self,name)
    self.empid=empid
class Manager(Employee):
  def __init__(self,name,empid,dept):
    Employee.__init__(self,name,empid)
    self.dept=dept
class SalesManager(Manager):
  def __init__(self,name,empid,dept,target):
    Manager.__init__(self,name,empid,dept)
    self.target=target

s=SalesManager("person1",1,"Maths",4000)
print(s.__dict__)
```

Hierarchical Inheritance

```python
class Vehicle:
  def __init__(self,model,company):
    self.model=model
    self.company=company

class Cars(Vehicle):
  def __init__(self,seats ,model,company):
    Vehicle.__init__(self,model,company)
    self.seats=seats

class Bikes(Vehicle):
  def __init__(self,seats ,model,company):
    Vehicle.__init__(self,model,company)
    self.seats=seats
```

Multiple Inheritance

```python
class Teacher:
  def __init__(self,name,subject):
    self.name=name
    self.subject=subject

class LabAssitant:
  def __init__(self,name,lab):
    self.name=name
    self.lab=lab

class TeacherLabAssitant(Teacher,LabAssitant):
  def __init__(self,name,subject,lab):
    Teacher.__init__(self,name,subject)
    LabAssitant.__init__(self,name,lab)
```
