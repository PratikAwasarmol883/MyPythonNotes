**FILTER**

**THERE IS ONLY ONE DIFFERENCE BETWEEN MAP AND FILTER.MAP WILL GIVE YOU THE PROCESSED VALUE AND FILTER WILL GIVE YOU THE ORIGINAL VALUE**

```python
def is_even(number):
  return number%2==0

def function7():
  l=[1,2,3,4,5,6,7,8,9]

  #return original value if the function returns TRUE
  evenno=list(filter(is_even,l))
  print(f"original values ={l}")
  print(evenno)
function7()
```

```python
def function8():
  l=[1,2,3,4,5,6,7,8,9]
  is_odd=list(filter(lambda x:x % 2 !=0,l))
  print(is_odd)
function8()
```

```python
#combining map and filter together
def function9():
  l=[1,2,3,4,5,6,7,8,9]
  print(f"{l}")

  #squares
  square=list(map(lambda x:x**2,l))
  print(f"{square}")

  #even
  even=list(filter(lambda x:x % 2 ==0,l))
  print(f"{even}")

  #square of even
  squareofeven=list(map(lambda x:x**2,even))
  print(f"{squareofeven}")

  #cube of odd Numbers
  oddno=list(filter(lambda x:x % 2 != 0,l))  #this will give us odd number
  print(oddno)

  cube=tuple(map(lambda x:x**3,oddno))  #this will give us odd number cube
  print(f"{cube}")

  #----------------------------------------------------
  cube1=list(map(lambda x:x**3,filter(lambda x:x % 2 != 0,l))) #we are putting filter inside lambda
  print(cube1)


function9()
```

#NESTED/INNER FUNCTION/LOCAL FUNCTION



1.   INNER FUNCTION IS LIKE VARIABLE TO OUTER FUNCTION
2.   SCOPE IS LOCAL
3.   YOU CAN HAVE AS MANY INNER FUNCTION AS YOU CAN NO LIMIT
4.   INNER FUNCTION CAN ACCESS ALL THE PROPERTIES/FUNCTIONS WHICH ARE DECLARE IN OUTER FUNCTION
5.   OUTER FUNCTION CANNOT ACCESS  THE PROPERTIES/FUNCTIONS WHICH ARE DECLARE IN INNER FUNCTION
6.   INNER FUNCTION CAN ACCESS OTHER INNER FUNCTION.




```python
def outerFunction():
  num=100
  print("Inside Outer Function")
  #local scope
  def innerfunction1():
    print("Inside Inner Function")
  print(f"Inner Function1 {innerfunction1},type={type(innerfunction1)}")
  innerfunction1()

  def innerfunction2():
    print("Inside InnerFunction2")
    innerfunction1()
    outerFunction()

outerFunction()
```

**NATURALLY WE CANNOT CALL INNER FNT OUTSIDE THE FUNCTION**

```python
def outer():
  print("Inside outer")
  num=100
  print(f"Before calling inner num={num}")
  def inner():
    print("Inside inner")
    num=300    #here inner function will create new local variable num
    print(f"inner num={num}")

  inner()
  print(f"After calling inner num={num}")

outer()
```

```python
def outer():
  print("Inside outer")
  num=100
  print(f"Before calling inner num={num}")
  def inner():
    print("Inside inner")

    #it will not create a local variable here
    nonlocal num

    num=300    #here inner function will create new local variable num
    #it will reassign outer num=300
    print(f"inner num={num}")

  inner()
  print(f"After calling inner num={num}")

outer()
```

**Global and Non-Local both are different.Local are local variable declaired inside outer function for global it will not be declare inside any function.**

#CLOSURE

WE CAN CALL INNER FUNCTION OUTSIDE OF OUTER FUNCTION BY RETURNING REFERENCE OF INNER FUNCTION BY OUTER FUNCTION.

closure.png

```python
def outer():
  print("Inside outer")

  def inner():
    print("inside Inner")

  return inner

result=outer()
print(f"Value of result={result},type={type(result)}")
result()
```

```python
#create table 4

def table(number):

  def inner():
    print("Inside Inner function")
    print([x*number for x in range(1,11)])

  return inner

table_4=table(4)     #table will return inners reference which will be stored in table_4
table_4()
```

#DECORATOR

```python
#DECORATOR

def decorator_result(func):
  def inner(p1,p2):
    print("------Decorator-------")
    func(p1,p2)
    print("-------"*20)
  return inner

@decorator_result
def add(p1,p2):
  print(f"Addition is : {p1+p2}")

@decorator_result
def sub(p1,p2):
  print(f"Substaction is : {p1-p2}")

@decorator_result
def multi(p1,p2):
  print(f"multiplication is : {p1*p2}")

@decorator_result
def division(p1,p2):
  print(f"Division is : {p1/p2}")


add(10,20)
sub(10,20)
multi(10,20)
division(10,20)
```

DECORATOR.png

```python
#outer function
def executor(func):
  print("Inside Executor")
  def wrapper():
    print("----"*20)
    func()
    print("----"*20)
  return wrapper

@executor
def func1():
  print("Inside Fuct 1")

func1()
```

JAVAS ANNOTATION IS EQUAL TO DECORATOR IN PYTHON

```python
#logger function

def log(func):        #-->Normal Decorator function
  def inner(p1,p2):
    result=func(p1,p2)
    print(f"Return value is: {result}")
  return inner

@log
def add(p1,p2):
  addition=p1+p2
  print(f"Addition is :{addition}")
  return addition

@log
def divide(p1,p2):
  division=p1/p2
  print(f"Division is :{division}")
  return division

add(10,20)
divide(9,3)
```

```python
#logger function

def log(func):        #-->Normal Decorator function
  def inner(*args,**kwargs):
    print(f"args ={args},kwargs ={kwargs}")
    result=func(*args,**kwargs)
    print(f"Return value is: {result}")
  return inner

@log
def add(p1,p2):
  addition=p1+p2
  print(f"Addition is :{addition}")
  return addition

@log
def divide(p1,p2):
  division=p1/p2
  print(f"Division is :{division}")
  return division

@log
def func3():
  print(f"Function 3")

add(10,20)
divide(9,3)
func3()
```

#PARAMETERIZED DECORATOR

IN EXAMPLE WE WANT TO CALL FUNCTION1 3 TIMES AND FUCTION2 5 TIMES.

```python
def repeat(count):    #decorator

  def repeat_decorator(func):

    def inner(*args,**kwargs):
      for index in range(count):
        func(*args,**kwargs)
    return inner
  return repeat_decorator

@repeat(count=3)
def function1():
  print("Inside function1")


@repeat(count=5)
def function2():
  print("Inside function2")

function1()
print("-*"*20)
function2()
```
